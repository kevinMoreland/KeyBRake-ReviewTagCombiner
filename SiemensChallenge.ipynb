```
def get_keywords(text):
  rake = Rake()
  rake.extract_keywords_from_text(text)
  return rake.get_ranked_phrases_with_scores()
```

```
json_data['keywords'] = json_data.apply(lambda row: get_keywords(row['reviewText']), axis = 1)
```
